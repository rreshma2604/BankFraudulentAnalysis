```python
import load_dataset_module
import user_statistics_module
def test_module():
    dic = load_dataset_module.makeDictionary()
    selected =0
    startBold = "\033[1m"
    endBold = "\033[0;0m"
    newline = "\n"
    print("**************************************************************************\n\t"+startBold
          +"\x1b[34mWelcome to Transaction fraudulent Detector Software\x1b[0m\n"+endBold
          +"**************************************************************************")
    print("**NOTE: To exit the software click enter button")
    
    #while loop to loop through the options
    while True:
        selected =0
        try:
            print("**************************************************************************")
            selected = int(input("Hi,Please select any one option from (1 to 8):"+newline
                                 +"1: Find the minimum and maximum of transactions of a user"+newline
                                 +"2: Location centroid of the user based on their transaction locations"+newline
                                 +"3: Distance of any transaction location from the centroid"+newline
                                 +"4: The standard deviation of transactions of any user"+newline
                                 +"5: Variance of user’s transactions"+newline
                                 +"6: Fraudulent status of a transaction"+newline
                                 +"7: Distance between any two given transactions of a user"+newline
                                 +"8: Distance of transactions of any two users"+newline
                                 +"9:"+startBold+" To Exist from software"+endBold+"\n"))
            
            if selected == 1: 
                print("You have selected option 1")
                uid = user_statistics_module.getUserId(dic)
                if(uid !=0):
                    minMax = user_statistics_module.minAndMax(dic[uid])
                    print(user_statistics_module.successMsg("The minimum transaction = "+str(minMax[0])+"\nThe maximum transaction= "+str(minMax[1])))
            elif selected == 2: 
                print("You have selected 2")
                uid = user_statistics_module.getUserId(dic)
                if(uid !=0):
                    print(user_statistics_module.successMsg("centroid = "),user_statistics_module.calcCentroid(dic[uid]))
            elif selected == 3: 
                print("You have selected 3")
                uid = user_statistics_module.getUserId(dic)
                if(uid !=0):
                    centroidxy = user_statistics_module.calcCentroid(dic[uid])
                    tid = user_statistics_module.getTrans(dic[uid])
                    if(tid !=0):
                        print(user_statistics_module.successMsg("Transaction loaction from centroid="), user_statistics_module.transLocation(dic[uid][tid],centroidxy))
            elif selected == 4: 
                print("you have selected 4")
                uid = user_statistics_module.getUserId(dic)
                if(uid !=0):
                    stdDev = user_statistics_module.StandardDeviation(dic[uid])
                    print(user_statistics_module.successMsg("standard deviation ="),stdDev[1])
            elif selected == 5: 
                print("you have selected 5")          
                uid = user_statistics_module.getUserId(dic)
                if(uid !=0):
                    stdDev = user_statistics_module.StandardDeviation(dic[uid])
                    print(user_statistics_module.successMsg("variance="), stdDev[0])
            elif selected == 6: 
                print("you have selected 6")
                uid = user_statistics_module.getUserId(dic)
                if(uid !=0):
                    tid = user_statistics_module.getTrans(dic[uid])
                    if(tid !=0):
                        print(user_statistics_module.identifyFraud(dic[uid][tid]))
            elif selected == 7: 
                print("you have selected 7")
                uid = user_statistics_module.getUserId(dic)
                if(uid !=0):
                    tid1 = user_statistics_module.getTrans(dic[uid])
                    if(tid1 !=0):
                        print("Insert the second transaction id")
                        tid2 = user_statistics_module.getTrans(dic[uid])
                        if(tid2 !=0):
                            print(user_statistics_module.successMsg("distance between any two given transactions of any user"), user_statistics_module.distbtw2trans(dic[uid][tid1], dic[uid][tid2])) 

            elif selected == 8: 
                print("you have selected 8")
                uid1 = user_statistics_module.getUserId(dic)
                if(uid1 !=0):
                    tid1 = user_statistics_module.getTrans(dic[uid1])
                    if(tid1 !=0):
                        print("Insert the detains of second user")
                        uid2 = user_statistics_module.getUserId(dic)
                        if(uid2 !=0):
                            tid2 = user_statistics_module.getTrans(dic[uid2])
                            if(tid2 !=0):
                                print(user_statistics_module.successMsg("the distance of transactions of any two users"), user_statistics_module.distanceBtwTransaction(dic[uid1][tid1],dic[uid2][tid2]))

            elif selected == 9:
                print(user_statistics_module.successMsg("you have selected to exit the software"))
                break
                
        except(ValueError): 
            print(user_statistics_module.errorMsg("Sorry but invalid value is obtained"))  
        except(KeyError):
            print(user_statistics_module.errorMsg("Sorry but invalid Key value is obtained"))
        except IOError:
            print("File not found")
        except AttributeError:
            print("NoneType object has no attribute")
        
        if selected > 9:
            print(user_statistics_module.errorMsg("You have inserted a wrong selection number"))
        if not selected:
            break
        print("**************************************************************************")

    print(user_statistics_module.successMsg("You have exited the software. Hope you like our service. Have a nice day"))

test_module()
```
